```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
def load_data():
    heart=pd.read_csv('heart.csv')
    lung=pd.read_csv('lung.csv')
    db=pd.read_csv('diabetes.csv')
    kidney=pd.read_csv('kidney.csv')
    return heart,lung,db,kidney
```

```python
heart,lung,db,kidney = load_data()
```

```python
kidney
```

```python
lung.columns = lung.columns.str.lower().str.replace(' ','_')
heart.columns = heart.columns.str.lower().str.replace(' ','_')
kidney.columns = kidney.columns.str.lower().str.replace(' ','_')
db.columns = db.columns.str.lower().str.replace(' ','_')
```

```python
# COLUMNS TO KEEP

heart = heart[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall', 'output']]
lung = lung[['age', 'smoking', 'yellow_fingers', 'anxiety', 'peer_pressure', 'chronic_disease', 'fatigue_', 'allergy_', 'wheezing', 'alcohol_consuming', 'coughing', 'shortness_of_breath', 'swallowing_difficulty', 'chest_pain', 'lung_cancer']]
db = db[['pregnancies','glucose','bloodpressure','skinthickness','insulin','bmi','diabetespedigreefunction','age','outcome']]
```

```python
def show_data():
    print('HEART',heart.head())
    print('LUNG',lung.head())
    # print('KIDNEY',kidney.head())
    print('DIABETES',db.head())
show_data()
```

```python
kidney = kidney.dropna()
kidney
```

```python
columns_to_transform_for_lung = ['smoking','yellow_fingers','anxiety','peer_pressure','chronic_disease','fatigue_','allergy_','wheezing','alcohol_consuming','coughing','shortness_of_breath','swallowing_difficulty','chest_pain','lung_cancer']
lung[columns_to_transform_for_lung] = lung[columns_to_transform_for_lung].replace({1:0,2:1,'YES':1,'NO':0})
lung
```

```python
xlung = lung.iloc[:,:-1]
ylung = lung.iloc[:,-1]
xheart = heart.iloc[:,:-1]
yheart = heart.iloc[:,-1]
xdb = db.iloc[:,:-1]
ydb = db.iloc[:,-1]
```

```python
# HEART SCALING
heart_scaler = MinMaxScaler()
heart_scaled = heart_scaler.fit_transform(xheart)

# Convert the scaled array back to a DataFrame and preserve column names
xheart = pd.DataFrame(heart_scaled, columns=xheart.columns)

# Check the result
print(heart.head())
```

```python
#LUNG SCALING
lung_scaler = MinMaxScaler()
lung_scaled = lung_scaler.fit_transform(xlung)
xlung = pd.DataFrame(lung_scaled,columns=xlung.columns)
# print(lung.head())
```

```python
#DIABETEST SCALING
db_scaler = MinMaxScaler()
db_scaled = db_scaler.fit_transform(xdb)
xdb=pd.DataFrame(db_scaled,columns=xdb.columns)
# print(db.head())
```

```python
xtrainheart,xtestheart,ytrainheart,ytestheart = train_test_split(xheart,yheart,test_size=0.2,random_state=0)
xtrainlung,xtestlung,ytrainlung,ytestlung = train_test_split(xlung,ylung,test_size=0.2,random_state=0)
xtraindb,xtestdb,ytraindb,ytestdb = train_test_split(xdb,ydb,test_size=0.2,random_state=0)
```

```python
#LUNG ALGORITHM SELECTION 
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100,random_state=42),
    "SVM": SVC(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(max_iter=1000, random_state=42)
}

# Train, predict and evaluate each model
for name, model in models.items():
    model.fit(xtrainlung, ytrainlung)
    predictions = model.predict(xtestlung)
    accuracy = accuracy_score(ytestlung, predictions)
    print(f'{name} Accuracy: {accuracy}')
    # Precision, Recall, F1-Score
    precision = precision_score(ytestlung, predictions)
    recall = recall_score(ytestlung, predictions)
    f1 = f1_score(ytestlung, predictions)

    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1-Score: {f1}')
    print("Confusion Matrix:")
    print(confusion_matrix(ytestlung, predictions))
    print('-------------------------')
```

```python
#HEART ALGORITHM SELECTION
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=1000,random_state=42),
    "SVM": SVC(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(max_iter=1000, random_state=42)
}

# Train, predict and evaluate each model
for name, model in models.items():
    model.fit(xtrainheart, ytrainheart)
    predictions = model.predict(xtestheart)
    accuracy = accuracy_score(ytestheart, predictions)
    print(f'{name} Accuracy: {accuracy}')
    precision = precision_score(ytestheart, predictions)
    recall = recall_score(ytestheart, predictions)
    f1 = f1_score(ytestheart, predictions)

    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1-Score: {f1}')
    print("Confusion Matrix:")
    print(confusion_matrix(ytestheart, predictions))
    print('-------------------------')
```

```python
#DIABETES ALGORITHM SELECTION
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=1000,random_state=42),
    "SVM": SVC(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(max_iter=1000, random_state=42)
}

# Train, predict and evaluate each model
for name, model in models.items():
    model.fit(xtraindb, ytraindb)
    predictions = model.predict(xtestdb)
    accuracy = accuracy_score(ytestdb, predictions)
    print(f'{name} Accuracy: {accuracy}')
    precision = precision_score(ytestdb, predictions)
    recall = recall_score(ytestdb, predictions)
    f1 = f1_score(ytestdb, predictions)

    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1-Score: {f1}')
    print("Confusion Matrix:")
    print(confusion_matrix(ytestdb, predictions))
    print('-------------------------')
```
